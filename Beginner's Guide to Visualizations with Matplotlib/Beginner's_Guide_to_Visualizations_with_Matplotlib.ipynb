```python
import matplotlib.pyplot as plt
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

# Create a line plot
plt.plot(x, y)

# Show the plot
plt.show()
```

```python
# Adding labels and a title
plt.plot(x,y, label='Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My Awesome Plot')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Adding line styles, colors, and markers
plt.plot(x,y, linestyle = '--', color = 'g', marker = 'o', label = 'Data Point')


# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15, 27, 10, 35]

# Create a bar chart
plt.bar(categories, values)

# Adding labels and a title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Show the plot
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# Create a scatter plot
plt.scatter(x, y)

# Adding labels and a title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')

# Show the plot
plt.show()
```
