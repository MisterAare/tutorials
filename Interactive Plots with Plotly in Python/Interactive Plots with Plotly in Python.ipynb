```python
import plotly.express as px
import pandas as pd

# Your BTC data
df = pd.read_csv('/content/BTC.csv')
data = df.head(40)
```

```python
fig = px.line(data, x="Date", y="Close", title="Sales Trend Over Time")
```

```python
fig.update_traces(hovertemplate="Date: %{x}<br>Close Price: $%{y:.2f}")
fig.update_layout(
    xaxis=dict(title="Date"),
    yaxis=dict(title="Close Price ($USD)"),
    xaxis_range=["2021-07-05", "2021-10-11"],  # Initial x-axis range
    dragmode="zoom",                           # Enable zooming & panning
)
```

```python
dropdown_buttons = [
    {"label": column, "method": "update", "args": [{"y": [data[column]], "yaxis.title.text": column}]}
    for column in data.columns[1:]
]
updatemenus = [{"buttons": dropdown_buttons, "direction": "down", "showactive": True}]
fig.update_layout(updatemenus=updatemenus)
```

```python
fig.show()
```
